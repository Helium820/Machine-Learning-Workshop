```python
import torch
```

# CPU

```python
# scalar mulitplication
%%timeit -n 100
x = torch.randn(1).to("cpu")
result = torch.matmul(x,x)
del x, result
```

```python
# matrix multiplication
%%timeit -n 1
x = torch.randn(10000,10000).to("cpu")
result = torch.matmul(x,x)
del x, result
```

# GPU

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
# scalar mulitplication
%%timeit -n 100
x = torch.randn(1,1).to(device)
result = torch.matmul(x,x)
del x, result
```

```python
# matrix multiplication
%%timeit -n 1
x = torch.randn(10000,10000).to(device)
result = torch.matmul(x,x)
del x, result
```
